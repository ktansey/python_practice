### Do students in more popular majors make more money?

```python
recent_grads.plot(x='Total', y='Median', kind='scatter')
```

Total is the total number of students (Men and Women) enrolled on a course, with more people enrolled on a course, the more popular it is. We see that in general the most popular course does not lead to a higher median salary, whereas for the least popular courses you may have the highest median salary and you may have the lowest. 

### Do students that majored in subjects that were majority female make more money?

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
```

There does seem to be a trend that the higher proportion of women on a course the lower the median salary. This could be because women are paid less than men, so more women in a field/area may result in a skew towards lower earnings.

## Grouped bar plot

Use a grouped bar plot to compare the number of men with the number of women in each category of majors.

```python
data = recent_grads.groupby(['Major_category'])[['Men', 'Women']].sum().reset_index()
    
N = len(data)

ind = np.arange(N)
width = 0.35  

p1 = plt.bar(ind, data['Men'], width, color='b')
p2 = plt.bar(ind + width, data['Women'], width, color='r')

plt.ylabel('Major Categories')
plt.title('Major Categories by gender')
plt.xticks(ind + width / 2, data['Major_category'], rotation = 90)
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```python
p1 = plt.bar(ind, data['Men'], width, color='b')
p2 = plt.bar(ind, data['Women'], width, color='r', bottom = data['Men'])

plt.ylabel('Major Categories')
plt.title('Major Categories by gender')
plt.xticks(ind, data['Major_category'], rotation = 90)
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

```python
data = recent_grads.groupby(['Major_category'])[['Men', 'Women', 'Total']].sum().reset_index()
data['Perc_Men'] = data['Men'] / data['Total']
data['Perc_Women'] = data['Women'] / data['Total']
```

```python
p1 = plt.bar(ind, data['Perc_Men'], width, color='b')
p2 = plt.bar(ind, data['Perc_Women'], width, color='r', bottom = data['Perc_Men'])

plt.ylabel('Major Categories')
plt.title('Major Categories by gender')
plt.xticks(ind, data['Major_category'], rotation = 90)
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

## Box plot

Use a box plot to explore the distributions of median salaries and unemployment rate.

```python
recent_grads.boxplot(column = 'Median', by = 'Major_category', rot = 90)
```

Engineering appears to have a higher median salary than other majors, with Psychology/Social Work having the lowest. 

```python
recent_grads[['Major_category', 'Median']].groupby('Major_category').median().reset_index().sort_values("Median")
```

Calculating the median of the median salaries, we see our conclusion from the plot is correct. 

```python
recent_grads.boxplot(column = 'Unemployment_rate', by = 'Major_category', rot = 90)
```

There is high unemployment in Law & Public Policy on average, but it is hard to see which major is performing the worst on average (the median). 

```python
recent_grads[['Major_category', 'Unemployment_rate']].groupby('Major_category').median().reset_index().sort_values("Unemployment_rate")
```

Social Science on average has the highest unemployment rate, with Education having the lowest. It would be interesting to know if employed individuals are actually working in their field of study.
